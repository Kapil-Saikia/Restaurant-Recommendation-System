```python
import pandas as pd

df = pd.read_csv(r"C:\Users\kapil\OneDrive\Documents\Dataset .csv")
df.head()
```

```python
df = df[['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Price range', 'Aggregate rating']]
df.dropna(inplace=True)
```

```python
def combine_features(row):
    return str(row['Cuisines']) + " " + str(row['Price range'])

df['combined_features'] = df.apply(combine_features, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combined_features'])

cosine_sim = cosine_similarity(count_matrix)
```

```python
user_input = "Chinese 2"
user_vec = cv.transform([user_input])
similarity_scores = cosine_similarity(user_vec, count_matrix)

# Get top 5 recommendations
import numpy as np

top_indices = similarity_scores[0].argsort()[-5:][::-1]

print("Recommended Restaurants:")
for idx in top_indices:
    print(df.iloc[idx]['Restaurant Name'], "| Cuisine:", df.iloc[idx]['Cuisines'], "| Rating:", df.iloc[idx]['Aggregate rating'])
```
